```python
%%sql

CREATE TABLE games_new (
    TEAM_ID INT,
    SEASON_ID INT,
    STANDINGSDATE DATE,
    CONFERENCE VARCHAR(10),
    TEAM VARCHAR(50),
    G INT,
    W INT,
    L INT,
    W_PCT FLOAT,
    HOME_RECORD VARCHAR(10),
    ROAD_RECORD VARCHAR(10)
);
```

```python
%%sql

SELECT * FROM games_new
```

```python
%%sql

INSERT INTO games_new (
    TEAM_ID,
    SEASON_ID,
    STANDINGSDATE,
    CONFERENCE,
    TEAM,
    G,
    W,
    L,
    W_PCT,
    HOME_RECORD,
    ROAD_RECORD)

SELECT TEAM_ID,
        SEASON_ID,
        STANDINGSDATE,
        CONFERENCE,
        TEAM,
        G,
        W,
        L,
        W_PCT,
        HOME_RECORD,
        ROAD_RECORD FROM ranking 
        WHERE CAST(SUBSTR(HOME_RECORD, 1, INSTR(HOME_RECORD, '-') - 1) AS INTEGER) > 30
        AND SUBSTR(STANDINGSDATE, 1, 4) BETWEEN '2021' AND '2022';
```

```python
%%sql

SELECT * FROM games_new
```

```python
%%sql

ALTER TABLE games_new ADD COLUMN Performance FLOAT;
```

```python
%%sql

UPDATE games_new
SET Performance = (W - CAST(SUBSTR(HOME_RECORD, 1, INSTR(HOME_RECORD, '-') - 1) AS INTEGER)) / W * 100;
```
